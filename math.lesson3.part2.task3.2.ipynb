```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plot
```

```python
plot.axes(projection='polar')
```

```python
rads = np.arange(0, (2*np.pi), 0.01)
```

```python
for radian in rads:
    plot.polar(radian,2,'o')
```

```python
plot.show()
```
