```python
# Почему прямые не кажутся перпендикулярными?

# Ответ: из-за координатной сетки.
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# оригинальный код

x = np.linspace(-5, 5, 21)
y = 3*x+1
y2 = (-1/3)*x+1
plt.plot(x,y)
plt.plot(x,y2)
plt.xlabel("x")
plt.ylabel("y")
```

```python
# "исправленный" код

plt.figure(figsize = (5, 5))
x = np.linspace(-15, 15)
y = 3*x+1
y2 = (-1/3)*x+1
plt.plot(x,y)
plt.plot(x,y2)
plt.xlabel("x")
plt.ylabel("y")
plt.ylim(-15, 15)
plt.xlim(-15, 15)
```
