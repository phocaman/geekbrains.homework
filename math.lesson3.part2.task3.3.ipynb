```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 5, 100)
y1 = 2*x
y2 = y1*0.5
plt.plot(x, y1, color='red')
plt.plot(x, y2, color='green')
```

```python
x = np.linspace(0, 5, 100)
y1 = 2*x
y2 = y1*0.5
plt.polar(x, y1, color='red')
plt.polar(x, y2, color='green')
```
